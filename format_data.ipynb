```python
import pandas as pd
import re
import os

# Path to the folder containing the image files
folder_path = 'datasets/Image_train'

# Initialize lists to store the extracted information
img_names = []
sex_values = []
age_values = []
height_values = []
weight_values = []

# Iterate over the files in the folder
for filename in os.listdir(folder_path):
    if filename.endswith('.jpg'):
        # Use regular expressions to extract information from the filename
        ret = re.match(r"\d+?_([FMfm])_(\d+?)_(\d+?)_(\d+).+", filename)
        if ret:
            sex = 0 if (ret.group(1) == 'F' or ret.group(1) == 'f') else 1
            age = int(ret.group(2))
            height = int(ret.group(3)) / 100000
            weight = int(ret.group(4)) / 100000

            # Append the extracted values to the respective lists
            img_names.append(filename)
            sex_values.append(sex)
            age_values.append(age)
            height_values.append(height)
            weight_values.append(weight)

# Create a Pandas DataFrame from the lists
data = {
    'img_name': img_names,
    'sex': sex_values,
    'age': age_values,
    'height': height_values,
    'weight': weight_values
}

df = pd.DataFrame(data)

# Display the DataFrame
print(df)
```
